```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
emperors = np.load('../Data/Emperors.npy').item()
```

```python
df = pd.DataFrame(columns=['emperor', 'start', 'end'])
for emperor, years in emperors.iteritems():
    df = df.append({'emperor': emperor, 
                'start': int(years[0]), 
                'end': int(years[1])}, 
             ignore_index=True)
```

```python
df
```

```python
conn = sqlite3.connect('../Data/Roman_Imperial_Coinage.sqlite')
cur = conn.cursor()
```

```python
cur.execute('''Create TABLE emperors(emperor TEXT, start INTEGER, end INTEGER)''')
cur.executemany("""INSERT INTO emperors(emperor, start, end) VALUES(?,?,?)""", 
                list(df.to_records(index=False)))
```

```python
conn.commit()
```
