```python
import pandas as pd
import requests
import io
import sys

stdout = sys.stdout
reload(sys)
sys.setdefaultencoding('utf-8')
sys.stdout = stdout
```

```python
url=("http://numismatics.org/ocre/query.csv?start=")
    
batch = url + "0"
rawContent = requests.get(batch).content
data = pd.read_csv(io.StringIO(rawContent.decode('utf-8')))
```

```python
data.tail()
```

```python
for i in range(1, 63):
    i *= 1000
    batch = url + str(i)
    rawContent = requests.get(batch).content
    df = pd.read_csv(io.StringIO(rawContent.decode('utf-8')))
    data = data.append(df, ignore_index=True)
```

```python
data.to_csv("../Data/OCRE.csv")
```
